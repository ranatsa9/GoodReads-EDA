```python
def clean_genres(value):
    genres = value.split(",")
    cleaned_genres = []

    for genre in genres:
        genre = genre.strip()
        parts = genre.split()

        genre_name = " ".join(parts[:-1])

        cleaned_genres.append(genre_name)

    return ", ".join(cleaned_genres)
```

```python
clean_df["genres"] = clean_df["genre_and_votes"].apply(
    clean_genres
)
```

```python
raw_df = pd.read_csv("goodreads_books.csv")

def get_year(value):
    if pd.isnull(value):
        return np.nan

    parts = str(value).replace("-", " ").replace("/", " ").split()

    for part in parts:
        if part.isdigit() and len(part) == 4:
            return int(part)

    return np.nan

raw_df["publication_year_fixed"] = raw_df["date_published"].apply(get_year)

year_map = raw_df.set_index("id")["publication_year_fixed"]

clean_df["publication_year"] = clean_df["id"].map(year_map).astype("Int64")

clean_df.to_csv("clean_goodreads_books.csv", index=False)
```

```python
high_rating = clean_df["average_rating"].quantile(0.75)
low_rating = clean_df["average_rating"].quantile(0.25)

minimum_popularity = clean_df["rating_count"].quantile(0.25)
median_popularity = clean_df["rating_count"].median()
high_popularity = clean_df["rating_count"].quantile(0.75)
```

```python
hidden_gems = clean_df[
    (clean_df["average_rating"] >= high_rating) &
    (clean_df["rating_count"] >= minimum_popularity) &
    (clean_df["rating_count"] < median_popularity)
]
```

```python
hidden_gems.shape
```

```python
hidden_gems[
    [
        "title",
        "author",
        "average_rating",
        "rating_count",
        "genres"
    ]
].sort_values(
    by="average_rating",
    ascending=False
).head(20)
```

```python
def classify_book(row):

    if (
        row["average_rating"] >= high_rating
        and row["rating_count"] >= minimum_popularity
        and row["rating_count"] < median_popularity
    ):
        return "Hidden Gem"

    elif (
        row["average_rating"] >= high_rating
        and row["rating_count"] >= median_popularity
    ):
        return "Popular and Highly Rated"

    elif (
        row["average_rating"] <= low_rating
        and row["rating_count"] >= high_popularity
    ):
        return "Overrated"

    elif row["rating_count"] < minimum_popularity:
        return "Under the Radar"

    else:
        return "Typical"
```

```python
clean_df["book_category"] = clean_df.apply(
    classify_book,
    axis=1
)
```

## Insights Summary

- Most books have average ratings close to 4.0, while very low ratings are uncommon.
- Popularity does not appear to have a strong relationship with average rating. Highly popular books are not always the highest rated.
- The most common genres have similar average ratings, although Nonfiction was slightly higher among the top genres.
- Rating counts are highly skewed, with a small number of books receiving millions of ratings.
- Book page counts contain several extreme outliers.
- Using rating and popularity percentiles, the books were grouped into categories such as Hidden Gem, Overrated, Popular and Highly Rated, and Under the Radar.
- The analysis identified 2,966 Hidden Gems and 2,413 Overrated books.
